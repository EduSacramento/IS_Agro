```python
# Importação de bibliotecas
import pandas as pd
import numpy as np
```

```python
# Variáveis utilizadas
dataset_name = 'risco_erosao_lavouras_temporarias'

raiz = '/home/eduardo/repositorios/IS_Agro/databases'
```

### Ouro

##### Lavouras Temporárias

```python
# Tabela de área colhida (ha) de lavouras temporárias
df_temp = (
    pd.read_parquet(f'{raiz}/prata_sidra_tab1612_lavoura_temporaria_area_colhida.parquet')
    .query('data >= 2000')
    .reset_index()
    .assign(geocodigo_uf = lambda df: df.geocodigo// 100000)
    .groupby(['geocodigo_uf', 'data', 'cultura'])
    .agg({'area':'sum'})
    .rename_axis(index={'geocodigo_uf': 'geocodigo'})
    .convert_dtypes()
)

# Tabela de área colhida (ha) da primeira safra de lavouras temporárias
df_ps = (
    pd.read_parquet(f'{raiz}/ouro_volume_agricola_lavouras_primeira_safra.parquet')
    .assign(geocodigo_uf = lambda df: df.geocodigo// 100000)
    .groupby(['geocodigo_uf', 'data', 'cultura'])
    .agg({'valor':'sum'})
    .rename_axis(index={'geocodigo_uf': 'geocodigo'})
    .convert_dtypes()
)

# Tabela de área colhida (ha) da segunda safra de lavouras temporárias
df_ss = (
    pd.read_parquet(f'{raiz}/ouro_volume_agricola_lavouras_segunda_safra.parquet')
    .assign(geocodigo_uf = lambda df: df.geocodigo// 100000)
    .groupby(['geocodigo_uf', 'data', 'cultura'])
    .agg({'valor':'sum'})
    .rename_axis(index={'geocodigo_uf': 'geocodigo'})
    .convert_dtypes()
)

# Dados sobre área (por Mha) de Sistema Plantio Direto a nível estadual
df_spd = (
    pd.read_parquet(f'{raiz}/prata_SPD.parquet')
    .reset_index()
)
```

```python
# Soja == 2713
df_soja = (
    df_temp
    .query('cultura == 2713 & area != 0')
    .reset_index()
    .drop(columns='cultura')
    .merge(
        df_ss
        .query('cultura == 2711 & valor != 0') # Dados da 2a. safra do milho
        .reset_index()
        .drop(columns='cultura'),
        how='left',
        on=['geocodigo', 'data']
    )
    .assign(
        area_SPD = lambda ha: ((ha.area / (ha.area + (ha.valor))) * ha.area)
    )
    .assign(**{
        'Soja - Convencional - Monocultura': lambda ha: round(((ha.area - ha.area_SPD) / 10000) * 0.5, 3), # 50% área soja - área soja SPD
        'Soja - Convencional - Sucessão': lambda ha: round(((ha.area - ha.area_SPD) / 10000) * 0.5, 3), # 50% área soja - área soja SPD
        'Soja - Plantio Direto': lambda ha: round((ha.area_SPD / 10000) * 0.7, 3), # 70% área soja SPD
        'Soja - SPD bem feito': lambda ha: round((ha.area_SPD / 10000) * 0.3, 3) # 30% área soja SPD
    })
    .drop(columns=(['area', 'valor', 'area_SPD']))
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        risco_erosivo = lambda df: np.where(df.cultura == 'Soja - Convencional - Monocultura', 8,
                                        np.where(df.cultura == 'Soja - Convencional - Sucessão', 5.7,
                                                 np.where(df.cultura == 'Soja - Plantio Direto', 1.8,
                                                          np.where(df.cultura == 'Soja - SPD bem feito', 0.3,
                                                                   pd.NA))))
    )
    .convert_dtypes()
)
```

```python
# Milho == 2711
df_milho = (
    df_ps
    .query('cultura == 2711 & valor != 0')
    .reset_index()
    .assign(**{
        'Milho - Monocultura (1a. safra)': lambda ha: round((ha.valor / 10000) * 0.8, 3), # (27% x 25%, 80% de área)
        'Milho (1a. safra)': lambda ha: round((ha.valor / 10000) * 0.2, 3), # (27% x 75%, 20% de área)
        # 'Milho - 2a. safra (verão)': lambda x: round((x.area/10000)*0.16, 3), # (73% x 20%, 16% de área)
        # 'Milho - 2a. safra + braquiária': lambda x: round((x.area/10000)*0.3, 3), # (73% x 40%, 30% de área)
        # 'Milho - 2a. safra - SPD': lambda x: round((x.area/10000)*0.3, 3) # (73% x 40%, 30% de área)
    })
    .drop(columns=(['cultura', 'valor']))
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        risco_erosivo = lambda df: np.where(df.cultura == 'Milho - Monocultura (1a. safra)', 57.3,
                                        np.where(df.cultura == 'Milho (1a. safra)', 11.5,
                                                 pd.NA))
    )
    .convert_dtypes()
)
```

```python
# Trigo == 2716
df_trigo = (
    df_ss
    .query('cultura == 2716')
    .reset_index()
    .assign(**{
        'Trigo - Convencional (2a. safra)': lambda ha: round((ha.valor / 10000) * 0.1, 6), # 10% de área
        'Trigo - Plantio Direto (no-till - 2a. safra)': lambda ha: round((ha.valor/ 10000) * 0.5, 6), # 50% de área
        'Trigo - Cultivo mínimo, SPD bem feito (2a. safra)': lambda ha: round((ha.valor / 10000) * 0.4, 6) # 40% de área
    })
    .drop(columns=(['cultura', 'valor']))
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        risco_erosivo = lambda df: np.where(df.cultura == 'Trigo - Convencional (2a. safra)', 5.3,
                                        np.where(df.cultura == 'Trigo - Plantio Direto (no-till - 2a. safra)', 1.5,
                                                 np.where(df.cultura == 'Trigo - Cultivo mínimo, SPD bem feito (2a. safra)', 0.3,
                                                          pd.NA)))
    )
    .convert_dtypes()
)
```

```python
# Algodão Herbáceo == 2689
df_algodao_herb = (
    df_ps
    .query('cultura == 2689')
    .reset_index()
    .assign(**{
        'Algodão - Convencional - morro abaixo': lambda ha: round((ha.valor / 10000) * 0.5, 9), # 40% de área
        'Algodão - Plano/sondulado, conservado': lambda ha: round((ha.valor / 10000) * 0.5, 9) # 60% de área
    })
    .drop(columns=(['cultura', 'valor']))
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        risco_erosivo = lambda df: np.where(df.cultura == 'Algodão - Convencional - morro abaixo', 26.0,
                                        np.where(df.cultura == 'Algodão - Plano/sondulado, conservado', 11.5,
                                                 pd.NA))
    )
    .convert_dtypes()
)
```

```python
# Cana-de-açúcar == 2696
df_cana = (
    df_temp
    .query('cultura == 2696 & area != 0')
    .reset_index()
    .assign(**{
        'Cana - Cana-planta somente': lambda ha: round((ha.area / 10000) * 0.2, 4), # 20% de área
        'Cana - Cana-planta + 4 cana-soca (5 colheitas)': lambda ha: round((ha.area / 10000) * 0.8, 4) # 80% de área
    })
    .drop(columns=(['cultura', 'area']))
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        risco_erosivo = lambda df: np.where(df.cultura == 'Cana - Cana-planta somente', 39.0,
                                        np.where(df.cultura == 'Cana - Cana-planta + 4 cana-soca (5 colheitas)', 8.3,
                                                 pd.NA))
    )
    .convert_dtypes()
)
```

```python
# Mandioca == 2708
df_mandioca = (
    df_temp
    .query('cultura == 2708 & area != 0')
    .reset_index()
    .assign(**{
        'Mandioca - Convencional - solo coberto (entressafra)': lambda ha: round(ha.area / 10000, 4)
    })
    .drop(columns=(['cultura', 'area']))
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        risco_erosivo = lambda df: np.where(df.cultura == 'Mandioca - Convencional - solo coberto (entressafra)', 0.4,
                                        pd.NA)
    )
    .convert_dtypes()
)
```

```python
# Arroz == 2692
df_arroz = (
    df_temp
    .query('cultura == 2692 & area != 0')
    .reset_index()
    .assign(**{
        'Arroz irrigado': lambda ha: np.round(np.where(ha.geocodigo.isin([42, 43]), ha.area/10000, 0), 2)
    })
    .drop(columns=(['cultura', 'area']))
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        risco_erosivo = lambda df: np.where(df.cultura == 'Arroz irrigado', 0,
                                        pd.NA)
    )
    .convert_dtypes()
)
```

```python
df_cereais = (
    df_ps
    .query('valor != 0')
    .reset_index()
    .groupby(['geocodigo', 'data'])
    .agg({'valor':'sum'})
    .reset_index()
    .merge(
        df_ps
        .query('cultura == 2713 & valor != 0'), # Soja 1a. safra
        how='left',
        on=['geocodigo', 'data'],
        suffixes=['_total', '_soja']
    )
    .merge(
        df_ps
        .query('cultura == 2711 & valor != 0'), # Milho 1a. safra
        how='left',
        on=['geocodigo', 'data'],
        #suffixes=['_soja', '_milho']
    )
    .merge(
        df_arroz # Arroz irrigado
        .assign(valor = lambda x: x.valor * 10000) # Retornando valor para ha para equivaler aos demais
        .drop(columns=['cultura', 'risco_erosivo']),
        how='left',
        on=['geocodigo', 'data'],
        suffixes=['_milho', '_arroz']
    )
    .merge(
        df_temp
        .query('cultura == 2708 & area != 0'), # Mandioca
        how='left',
        on=['geocodigo', 'data'],
        #suffixes=['_total', '_mandioca']
    )
    .merge(
        df_temp
        .query('cultura == 2696 & area != 0'), # Cana-de-açúcar
        how='left',
        on=['geocodigo', 'data'],
        suffixes=['_mandioca', '_cana']
    )
    .merge(
        df_ps
        .query('cultura == 2698'), # Algodão 1a. safra
        how='left',
        on=['geocodigo', 'data'],
        #suffixes=['_arroz', '_algodao']
    )
    .fillna(0)
    .assign(**{
        'cereais_total': lambda n: n.valor_total - (n.valor_soja + n.valor_milho + n.valor_arroz + n.area_mandioca + n.area_cana + n.valor),
        'Cereais - Convencional, sem cobertura': lambda x: x.cereais_total * 0.3,
        'Cereais - Convencional': lambda x: x.cereais_total * 0.4,
        'Cereais - Sistema Plantio Direto': lambda x: x.cereais_total * 0.3
    })
    .drop(columns={'valor_total', 'valor_soja', 'valor_milho', 'valor_arroz', 'area_mandioca', 'area_cana', 'valor', 'cereais_total'})
    .melt(
        id_vars=['geocodigo', 'data'],
        var_name='cultura',
        value_name='valor'
    )
    .assign(
        valor = lambda ha: round(ha.valor/10000, 2),
        risco_erosivo = lambda df: np.where(df.cultura == 'Cereais - Convencional, sem cobertura', 41.5,
                                        np.where(df.cultura == 'Cereais - Convencional', 4.9,
                                                 np.where(df.cultura == 'Cereais - Sistema Plantio Direto', 1.7,
                                                          pd.NA)))
        )
    .convert_dtypes()
    .reset_index(drop=True)
)
```

```python
df_final = (
    pd.concat([
        df_soja,
        df_milho,
        df_trigo,
        df_algodao_herb,
        df_cana,
        df_mandioca,
        df_arroz,
        df_cereais
    ])
    .sort_values(['geocodigo', 'data'])
    .reset_index(drop=True)
)
```

```python
df_ouro = (
    df_final
    .rename(columns={
        'geocodigo': 'state',
        'valor': 'value',
        'data': 'date',
        'cultura': 'label',
        'risco_erosivo': 'sub_label'
        })
    .assign(country = 76)
    .filter(['country', 'state', 'date', 'label', 'sub_label', 'value'])
)

df_ouro.to_csv(f'{raiz}/ouro_{dataset_name}.csv', index=False)
```
